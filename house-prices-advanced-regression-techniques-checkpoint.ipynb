# 2. Data cleaning

````python
c_train['train']  = 1
c_test['train']  = 0
df = pd.concat([c_train, c_test], axis=0,sort=False)
````

````python
#Percentage of NAN Values 
NAN = [(c, df[c].isna().mean()*100) for c in df]
NAN = pd.DataFrame(NAN, columns=["column_name", "percentage"])
````

````python
NAN = NAN[NAN.percentage > 50]
NAN.sort_values("percentage", ascending=False)
````

````python
#Drop PoolQC, MiscFeature, Alley and Fence features
df = df.drop(['Alley','PoolQC','Fence','MiscFeature'],axis=1)
````

````python
categorical = df.select_dtypes(include=['object'])
numeric =df.select_dtypes(exclude=['object'])
````

#### Dealing with categorical features

````python
#Number of null values in each feature
null_counts = categorical.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
````


We will fill  ```BsmtQual```, ```BsmtCond```,```BsmtExposure```,``` BsmtFinType1```,``` BsmtFinType2```, 

```GarageType```,``` GarageFinish```,``` GarageQual```,``` FireplaceQu```,``` GarageCond```  with "```None```" (Take a look in the data description).
 We will fill the rest of features with th most frequent value (using its own most frequent value).

````python
columns_None = ['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','GarageType','GarageFinish','GarageQual','FireplaceQu','GarageCond']
categorical[columns_None]= categorical[columns_None].fillna('None')
````

````python
columns_with_lowNA = ['MSZoning','Utilities','Exterior1st','Exterior2nd','MasVnrType','Electrical','KitchenQual','Functional','SaleType']
#fill missing values for each column (using its own most frequent value)
categorical[columns_with_lowNA] = categorical[columns_with_lowNA].fillna(categorical.mode().iloc[0])
````

#### Dealing with numeric features

````python
#Number of null values in each feature
null_counts = numeric.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
````

````python
print((numeric['YrSold']-numeric['YearBuilt']).median())
print(numeric["LotFrontage"].median())
````

So we will fill the year with 1979 and the Lot frontage with 68

````python
numeric['GarageYrBlt'] = numeric['GarageYrBlt'].fillna(numeric['YrSold']-35)
numeric['LotFrontage'] = numeric['LotFrontage'].fillna(68)
````

````python
numeric= numeric.fillna(0)
````

#### Delete features with low variance 

````python
categorical = categorical.drop(['Heating','RoofMatl','Condition2','Street','Utilities'],axis=1)
````

#### Create new features

````python
numeric['Age_House']= (numeric['YrSold']-numeric['YearBuilt'])
numeric['Age_House'].describe()
````

````python
Negatif = numeric[numeric['Age_House'] < 0]
Negatif
````

````python
numeric.loc[numeric['YrSold'] < numeric['YearBuilt'],'YrSold' ] = 2009
numeric['Age_House']= (numeric['YrSold']-numeric['YearBuilt'])
numeric['Age_House'].describe()
````

TotalBsmtBath : Sum of : BsmtFullBath and 1/2 BsmtHalfBath

TotalBath : Sum of : FullBath and 1/2 HalfBath

TotalSA : Sum of : 1stFlrSF and 2ndFlrSF and basement area </font

````python
numeric['TotalBsmtBath'] = numeric['BsmtFullBath'] + numeric['BsmtFullBath']*0.5
numeric['TotalBath'] = numeric['FullBath'] + numeric['HalfBath']*0.5 
numeric['TotalSA']=numeric['TotalBsmtSF'] + numeric['1stFlrSF'] + numeric['2ndFlrSF']
````

````python
numeric.head()
````

#### Now the next step is to encode categorical features
#### Ordinal categories features - Mapping from 0 to N

````python
bin_map  = {'TA':2,'Gd':3, 'Fa':1,'Ex':4,'Po':1,'None':0,'Y':1,'N':0,'Reg':3,'IR1':2,'IR2':1,'IR3':0,"None" : 0,
            "No" : 2, "Mn" : 2, "Av": 3,"Gd" : 4,"Unf" : 1, "LwQ": 2, "Rec" : 3,"BLQ" : 4, "ALQ" : 5, "GLQ" : 6
            }
categorical['ExterQual'] = categorical['ExterQual'].map(bin_map)
categorical['ExterCond'] = categorical['ExterCond'].map(bin_map)
categorical['BsmtCond'] = categorical['BsmtCond'].map(bin_map)
categorical['BsmtQual'] = categorical['BsmtQual'].map(bin_map)
categorical['HeatingQC'] = categorical['HeatingQC'].map(bin_map)
categorical['KitchenQual'] = categorical['KitchenQual'].map(bin_map)
categorical['FireplaceQu'] = categorical['FireplaceQu'].map(bin_map)
categorical['GarageQual'] = categorical['GarageQual'].map(bin_map)
categorical['GarageCond'] = categorical['GarageCond'].map(bin_map)
categorical['CentralAir'] = categorical['CentralAir'].map(bin_map)
categorical['LotShape'] = categorical['LotShape'].map(bin_map)
categorical['BsmtExposure'] = categorical['BsmtExposure'].map(bin_map)
categorical['BsmtFinType1'] = categorical['BsmtFinType1'].map(bin_map)
categorical['BsmtFinType2'] = categorical['BsmtFinType2'].map(bin_map)

PavedDrive =   {"N" : 0, "P" : 1, "Y" : 2}
categorical['PavedDrive'] = categorical['PavedDrive'].map(PavedDrive)
````

#### Will we use One hot encoder to encode the rest of categorical features

````python
#Select categorical features
rest_object_columns = categorical.select_dtypes(include=['object'])
#Using One hot encoder
categorical = pd.get_dummies(categorical, columns=rest_object_columns.columns) 
````

````python
categorical.head()
````

````python
df_final = pd.concat([categorical, numeric], axis=1,sort=False)
df_final.head()
````

````python
df_final = df_final.drop(['Id',],axis=1)

df_train = df_final[df_final['train'] == 1]
df_train = df_train.drop(['train',],axis=1)


df_test = df_final[df_final['train'] == 0]
df_test = df_test.drop(['SalePrice'],axis=1)
df_test = df_test.drop(['train',],axis=1)
````

````python
target= df_train['SalePrice']
df_train = df_train.drop(['SalePrice'],axis=1)
````

# 3. Machine Learning

````python
x_train,x_test,y_train,y_test = train_test_split(df_train,target,test_size=0.33,random_state=0)
````

````python
xgb =XGBRegressor( booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=0.6, gamma=0,
             importance_type='gain', learning_rate=0.01, max_delta_step=0,
             max_depth=4, min_child_weight=1.5, n_estimators=2400,
             n_jobs=1, nthread=None, objective='reg:linear',
             reg_alpha=0.6, reg_lambda=0.6, scale_pos_weight=1, 
             silent=None, subsample=0.8, verbosity=1)


lgbm = LGBMRegressor(objective='regression', 
                                       num_leaves=4,
                                       learning_rate=0.01, 
                                       n_estimators=12000, 
                                       max_bin=200, 
                                       bagging_fraction=0.75,
                                       bagging_freq=5, 
                                       bagging_seed=7,
                                       feature_fraction=0.4, 
                                       )
````

````python
#Fitting
xgb.fit(x_train, y_train)
lgbm.fit(x_train, y_train,eval_metric='rmse')
````

````python
predict1 = xgb.predict(x_test)
predict = lgbm.predict(x_test)
````

````python
print('Root Mean Square Error test = ' + str(math.sqrt(metrics.mean_squared_error(y_test, predict1))))
print('Root Mean Square Error test = ' + str(math.sqrt(metrics.mean_squared_error(y_test, predict))))
````

````python
xgb.fit(df_train, target)
lgbm.fit(df_train, target,eval_metric='rmse')
````

````python
predict4 = lgbm.predict(df_test)
predict3 = xgb.predict(df_test)
predict_y = ( predict3*0.45 + predict4 * 0.55)
````
